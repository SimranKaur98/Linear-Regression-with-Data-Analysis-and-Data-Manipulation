```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\hp\AppData\Local\Programs\Python\Python36\Lib\site-packages\sklearn\datasets\data\diamonds.csv",index_col=0)
data = data.sample(frac=0.2)
data.reset_index(inplace=True,drop=True)
data.head()
```

```python
data["clarity"].unique()
```

```python
#Replace the categorical data to numeric
clarity_dict = {'I1':0,'SI2':1, 'SI1':2, 'VS1':4, 'VS2':3, 'VVS2':5, 'VVS1':6, 'IF':7}
data["clarity"].replace(clarity_dict, inplace=True)
data.head()
```

```python
#Categorical data that is nominal or ordinal, one hot encoding is the good way to show categories in numeric form
data = pd.get_dummies(data,columns=["cut","color"])
data.head()
```

```python
NUMERIC_FEATURES = ['carat','depth','table','x','y','z']
numeric_df = data[NUMERIC_FEATURES]
numeric_df.describe()
```

```python
#standardize numeric data by subtracting the mean & dividing by the std dev
from sklearn.preprocessing import scale
numeric_array = scale(numeric_df)
numeric_array[:5]
```

```python
numeric_df = pd.DataFrame(numeric_array, columns=numeric_df.columns)
numeric_df.describe()
```

```python
#Replacing original data with the standardized data
data.drop(NUMERIC_FEATURES, inplace = True, axis = 1)
data.head()
```

```python
data = pd.concat([data,numeric_df],axis =1)
```

```python
data.shape
```

```python
data.head()
```

```python
#Saving preprocessed data into a csv file
data.to_csv(r"C:\Users\hp\AppData\Local\Programs\Python\Python36\Lib\site-packages\sklearn\datasets\data\diamonds_processed.csv", index=False)
```

```python
#Perform LR
from sklearn.model_selection import train_test_split
X = data.drop("price", axis =1)
Y = data["price"]
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_prediction = lr.predict(x_test)
y_prediction
```

```python
#Evaluation
train_score = lr.score(x_train, y_train)
train_score
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_prediction)
print("Testing Score : ",score)
```
