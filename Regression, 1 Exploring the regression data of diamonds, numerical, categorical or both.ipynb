```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\hp\AppData\Local\Programs\Python\Python36\Lib\site-packages\sklearn\datasets\data\diamonds.csv",index_col=0)
data.head()
```

```python
data.shape
```

```python
data = data.sample(frac=0.2)
```

```python
data.shape
```

```python
data.reset_index(inplace=True,drop=True)
```

```python
data.head()
```

```python
correlation = data.corr() 
correlation
```

```python
#heatmaps are the best way to visualize correlation between features
plt.subplots(figsize=(10,8))
sns.heatmap(correlation,annot=True)
plt.show()
```
